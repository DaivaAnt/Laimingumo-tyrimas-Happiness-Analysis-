```python
# Funkcija outlier'ų aptikimui naudojant IQR
def detect_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
    return outliers

# Taikomas outlier'ų aptikimas kiekvienam veiksniui
outliers = detect_outliers_iqr(final_data[['Log GDP per capita', 'Social support',
                                           'Healthy life expectancy', 'Freedom to make life choices',
                                           'Generosity', 'Perceptions of corruption']])
print("Outlier'iai pagal IQR metodą:\n", outliers)
```

```python
# Naudoju IQR metodą viršutinėms ir apatinėms riboms nustatyti
Q1 = final_data[['Social support', 'Generosity']].quantile(0.25)
Q3 = final_data[['Social support', 'Generosity']].quantile(0.75)
IQR = Q3 - Q1

# Apatinė ir viršutinė riba
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Taikau apribojimus (capping)
final_data['Social support'] = final_data['Social support'].clip(lower=lower_bound['Social support'], upper=upper_bound['Social support'])
final_data['Generosity'] = final_data['Generosity'].clip(lower=lower_bound['Generosity'], upper=upper_bound['Generosity'])

print("Outlier'iai apriboti naudojant capping metodą.")
```
